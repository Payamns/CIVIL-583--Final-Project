```python
## Parameters estimation
##(starting value. lower band , upper band, 0:estimate for each   1:keep this value fixed or reference level)
ASC_CAR = Beta('ASC_CAR', 0 , None, None, 0)
ASC_TRAIN = Beta('ASC_TRAIN', 0 , None, None, 0)
ASC_SM = Beta('ASC_SM', 0 , None, None, 1)
B_TIME = Beta('B_TIME', 0 , None, None, 0)
B_COST = Beta('B_COST', 0 , None, None, 0)
```

```python
## new variables definition
# add new column to database

CAR_AV_SP = DefineVariable('CAR_AV_SP', CAR_AV  * (SP != 0),database)
TRAIN_AV_SP = DefineVariable('TRAIN_AV_SP', TRAIN_AV  * (SP != 0),database)
TRAIN_TT_SCALED = DefineVariable('TRAIN_TT_SCALED', TRAIN_TT / 100.00 ,database)
TRAIN_COST_SCALED = DefineVariable('TRAIN_COST_SCALED', TRAIN_CO / 100.00 ,database)
SM_TT_SCALED = DefineVariable('SM_TT_SCALED', SM_TT/ 100.00 ,database)
SM_COST_SCALED = DefineVariable('SM_COST_SCALED', SM_CO/ 100.00 ,database)
CAR_TT_SCALED = DefineVariable('CAR_TT_SCALED', CAR_TT/ 100.00 ,database)
CAR_COST_SCALED = DefineVariable('CAR_COST_SCALED', CAR_CO/ 100.00 ,database)
```

```python
## Utility functions Specification
# train:1, CAR = 3,SWISSMETRO = 2 
v1 = ASC_TRAIN + B_TIME * TRAIN_TT_SCALED + B_COST * TRAIN_COST_SCALED
v2 = ASC_SM + B_TIME * SM_TT_SCALED + B_COST * SM_COST_SCALED
v3 = ASC_CAR + B_TIME * CAR_TT_SCALED + B_COST * CAR_COST_SCALED

#Utility of alternatives
v = {1: v1, 2:v2, 3:v3}
#Aavailability of altenatives
av = {1: TRAIN_AV_SP, 2:SM_AV, 3: CAR_AV_SP }
```

```python
##loglikelihood function
#(utility, availability, choice set)
logprob = models.loglogit(v,av,CHOICE)
```

```python
biogeme = bio.BIOGEME(database,logprob)
biogeme.modelName = "01logit"
#the name will be used to create output file
```

```python
results = biogeme.estimate()
```

```python
###trips are imprted from second step to be divided into mode share table
import numpy as np
trips = np.load('C:/Users/User/Desktop/583-chicago/2step/TripDistribution.npy')

```

```python
###Monte Carlo to divide Chicago trip distribution table 
import random
transit = np.zeros(trips.shape)
sov= np.zeros(trips.shape)
hov = np.zeros(trips.shape)
sov_range= 0.746
hov_range = 0.746 + 0.118
transit_range = 0.746 + 0.118 + 0.135

i = 0
j = 0
count = 0
for i in range(len(trips)):
    for j in range(len(trips)):
        count = 0
        while count < trips[i,j]:
            rand = random.uniform(0,1)
            if rand < sov_range:
                sov[i,j] += 1
            elif rand < hov_range:
                hov[i,j] += 1
            elif rand < transit_range:
                transit[i,j] += 1
            else:
                continue
        
            count += 1
    
```
