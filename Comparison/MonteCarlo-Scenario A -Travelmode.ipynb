```python
###trips are imprted from second step to be divided into mode share table
import numpy as np
trips = np.load('C:/Users/User/Desktop/583-chicago/2step/TripDistribution.npy')

###Monte Carlo to divide Chicago trip distribution table 
import random
transit = np.zeros(trips.shape)
sov= np.zeros(trips.shape)
hov = np.zeros(trips.shape)
other = np.zeros(trips.shape)
## boundries are ectracted from the figure in the report
sov_range= 0.796
hov_range = 0.796 + 0.168
transit_range = 0.796 + 0.168 + 0.035

i = 0
j = 0
count = 0
for i in range(len(trips)):
    for j in range(len(trips)):
        count = 0
        while count < trips[i,j]:
            rand = random.uniform(0,1)
            if rand < sov_range:
                sov[i,j] += 1
            elif rand < hov_range:
                hov[i,j] += 1
            elif rand < transit_range:
                transit[i,j] += 1
            else:
                other[i,j] += 1
        
            count += 1
    
## save results as csv files
    
from numpy import savetxt
savetxt("ScenarioA-SOV_trips.csv", sov, delimiter = ',')
savetxt("ScenarioA-HOV_trips.csv", hov, delimiter = ',')
savetxt("ScenarioA-Transit_trips.csv", transit, delimiter = ',')
savetxt("Other_trips.csv", other, delimiter = ',')
```

```python
###showing the progress of convergence
sov_share = []
hov_share= []
transit_share = []
sov_count= 0 
hov_count = 0
transit_count = 0
other_count = 0

sov_range= 0.746
hov_range = 0.746 + 0.118
transit_range = 0.746 + 0.118 + 0.135


    
count = 0
while count <1000:
    rand = random.uniform(0,1)
    if rand < sov_range:
        sov_count += 1
    elif rand < hov_range:
        hov_count += 1
    elif rand < transit_range:
        transit_count += 1
    else:
        other_count += 1
        
    count += 1    
        
    sov_share.append(sov_count/count)
    hov_share.append(hov_count/ count)
    transit_share.append(transit_count / count)
        
savetxt("ScenarioA-SOV_sahre.csv", sov_share, delimiter = ',')
savetxt("ScenarioA-HOV_sahre.csv", hov_share, delimiter = ',')
savetxt("ScenarioA-Transit_share.csv", transit_share, delimiter = ',')


```
